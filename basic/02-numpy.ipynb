## Linear Algebra

$$
A = 
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
$$

$$
det(A) = ad-bc
$$

```python
A = np.array([[1, 2], [3, 4]])
# 1*4 - 2*3 = -2
np.linalg.det(A)
```

$$
A = 
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
$$

$$
inv(A) = A^{-1} = \frac{1}{ad - bc}
\begin{bmatrix}
d & -b \\
-c & a
\end{bmatrix}

$$

```python
np.linalg.inv(A)
```

$$
A = 
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
$$

$$
A \cdot A = 
\begin{bmatrix}
a^2 + bc & ab + bd \\
ca + dc & cb + d^2
\end{bmatrix}
$$

```python
np.dot(A, A)
```
