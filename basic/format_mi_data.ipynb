## format detail data

```python
import json
from datetime import datetime, timedelta

# Load the original JSON
with open("../data/sleep_data.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# UTC+8 offset
utc8_offset = timedelta(hours=8)

# Helper to convert UNIX timestamp → UTC+8 string
def ts_to_utc8(ts: int) -> str:
    return (datetime.utcfromtimestamp(ts) + utc8_offset).strftime("%Y-%m-%d %H:%M:%S")

# Convert main sleep times
data["bedtime_utc8"] = ts_to_utc8(data.get("bedtime") or data.get("device_bedtime"))
data["wake_up_time_utc8"] = ts_to_utc8(data.get("wake_up_time") or data.get("device_wake_up_time"))

# Convert each stage
for stage in data.get("items", []):
    stage["start_time_utc8"] = ts_to_utc8(stage["start_time"])
    stage["end_time_utc8"] = ts_to_utc8(stage["end_time"])

# Save to new JSON file
with open("sleep_data_utc8.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)
```

## format aggragate data

```python
import json
from datetime import datetime, timedelta

# Load original JSON file
with open("../data/sleep_data_aggregated.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# UTC+8 offset
utc8_offset = timedelta(hours=8)

def ts_to_utc8(ts: int) -> str:
    """Convert UNIX timestamp to UTC+8 human-readable string"""
    return (datetime.utcfromtimestamp(ts) + utc8_offset).strftime("%Y-%m-%d %H:%M:%S")

# Convert main sleep times if present
if "bedtime" in data or "device_bedtime" in data:
    data["bedtime_utc8"] = ts_to_utc8(data.get("bedtime") or data.get("device_bedtime"))
if "wake_up_time" in data or "device_wake_up_time" in data:
    data["wake_up_time_utc8"] = ts_to_utc8(data.get("wake_up_time") or data.get("device_wake_up_time"))

# Convert 'items' timestamps if they exist
if "items" in data:
    for stage in data["items"]:
        stage["start_time_utc8"] = ts_to_utc8(stage["start_time"])
        stage["end_time_utc8"] = ts_to_utc8(stage["end_time"])

# Convert 'segment_details' timestamps if they exist
if "segment_details" in data:
    for segment in data["segment_details"]:
        if "bedtime" in segment:
            segment["bedtime_utc8"] = ts_to_utc8(segment["bedtime"])
        if "wake_up_time" in segment:
            segment["wake_up_time_utc8"] = ts_to_utc8(segment["wake_up_time"])

# Save to new JSON file
with open("sleep_data_aggregated_utc8.json", "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=2)
```

## convert to utc+8 time

```python
from datetime import datetime, timezone, timedelta

# Create a UTC+8 timezone object
UTC8 = timezone(timedelta(hours=8))

def to_utc8(ts: int) -> datetime:
    """
    Convert Unix timestamp (seconds) to UTC+8 datetime using timezone-aware objects.
    """
    # 1. Interpret timestamp as UTC (aware datetime)
    dt_utc = datetime.fromtimestamp(ts, timezone.utc)

    # 2. Convert to UTC+8 timezone
    dt_utc8 = dt_utc.astimezone(UTC8)

    return dt_utc8


# Example:

ts = 1764880260
print(to_utc8(ts))
```
