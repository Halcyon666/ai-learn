```python
def printRangeDoubleSquare(n):
    a = []
    for i in range(n):
        # a.append((i + 1)**2)
        a.append(pow(i+1, 2))
    return a
    
printRangeDoubleSquare(5)
    
```

### 列表推导式

```python
def printRangeDoubleSquare(n):
    print([pow(i, 2) for i in range(1, n+1)])
    
printRangeDoubleSquare(5)
```

[表达式 for 变量 in 可迭代对象 if 条件]

含义：

表达式：最终放进列表里的东西

for 变量 in 可迭代对象：循环部分，类似普通的 for 循环

if 条件（可选）：筛选哪些元素需要放进列表

```python
a = [i**2 for i in range (5, 10) if i % 2 == 0]
print(a)
```

# day2 正则替换重命名文档

```python
import os
import re

def rename_files_with_sidebar_position(root_dir):
    # 匹配 sidebar_position: 数字 \s* 0 个或多个空白字符
    pattern = re.compile(r"sidebar_position:\s*(\d+)")

    for dirpath, _, filenames in os.walk(root_dir):
        for filename in filenames:
            filepath = os.path.join(dirpath, filename)

            # 只处理文本类文件
            if not os.path.isfile(filepath):
                continue

            try:
                with open(filepath, "r", encoding="utf-8") as f:
                    content = f.read()
            except Exception as e:
                print(f"⚠️ 无法读取 {filepath}: {e}")
                continue

            match = pattern.search(content)
            if match:
                number = int(match.group(1))

                # 去掉已有的数字前缀，比如 "02-xxx.md"
                original_name = re.sub(r"^\d{2}-", "", filename)

                new_filename = f"{number:02d}-{original_name}"
                new_filepath = os.path.join(dirpath, new_filename)

                if filepath != new_filepath:
                    if not os.path.exists(new_filepath):
                        print(f"✅ 重命名: {filename} -> {new_filename}")
                        os.rename(filepath, new_filepath)
                    else:
                        print(f"⏩ 跳过: {new_filename} 已存在")
            else:
                print(f"❌ 未找到 sidebar_position: {filename}")


# if __name__ == "__main__":
    # folder = "D:/summary/docs"  # 你要处理的目录
    # rename_files_with_sidebar_position(folder)
```

```python
import os
import re

def clean_sidebar_position_line_with_adjacent_empty(root_dir):
    # 匹配 sidebar_position: 行，包含前后的空行
    # 逻辑：匹配前面零个或多个空行 + sidebar_position 行 + 后面零个或多个空行
    pattern = re.compile(r'\n?\s*sidebar_position:.*\n?', re.MULTILINE)

    for dirpath, _, filenames in os.walk(root_dir):
        for filename in filenames:
            filepath = os.path.join(dirpath, filename)

            if not os.path.isfile(filepath):
                continue

            try:
                with open(filepath, "r", encoding="utf-8") as f:
                    content = f.read()
            except Exception as e:
                print(f"⚠️ 无法读取 {filepath}: {e}")
                continue

            # 删除匹配行及紧邻的空行
            new_content = pattern.sub('\n', content)  # 用单个换行代替，避免连续空行过多

            # 去掉开头和末尾多余换行
            new_content = new_content.strip() + '\n'

            if new_content != content:
                try:
                    with open(filepath, "w", encoding="utf-8") as f:
                        f.write(new_content)
                    print(f"✅ 已删除 sidebar_position 行及其相邻空行: {filepath}")
                except Exception as e:
                    print(f"⚠️ 无法写入 {filepath}: {e}")

# if __name__ == "__main__":
#     folder = "D:/summary/docs"  # 你要处理的目录
#     clean_sidebar_position_line_with_adjacent_empty(folder)
```

```python
import os

def append_tail_protocol(root_dir):
    for dirpath, _, filenames in os.walk(root_dir):
        for filename in filenames:
            filepath = os.path.join(dirpath, filename)

            if not os.path.isfile(filepath):
                continue

            try:
                with open(filepath, "r", encoding="utf-8") as f:
                    content = f.read()
            except Exception as e:
                print(f"⚠️ 无法读取 {filepath}: {e}")
                continue

            tail_marker = "<TailProtocal />"
            new_content = content.rstrip() + "\n\n" + tail_marker + "\n"

            if content.strip().endswith(tail_marker):
                print(f"⏩ 跳过: 已有 {tail_marker} -> {filepath}")
                continue

            try:
                with open(filepath, "w", encoding="utf-8") as f:
                    f.write(new_content)
                print(f"✅ 已追加 {tail_marker}: {filepath}")
            except Exception as e:
                print(f"⚠️ 无法写入 {filepath}: {e}")


# if __name__ == "__main__":
#     folder = "D:/summary/docs"  # 你要处理的目录
#     append_tail_protocol(folder)
```

# day3 password gen

```python
import random
import string

def generate_random_code(length=15):
    # 可选字符集：大写、小写、数字、符号
    
    chars = string.ascii_letters + string.digits + string.punctuation
    # 从字符集中随机选取 length 个字符
    return ''.join(random.choice(chars) for _ in range(length))

print(generate_random_code())
```

# day4 hash chain

```python
import hashlib
prev = '0000'
for i in range(3):
    data = f"Block {i} prev={prev}"
    h = hashlib.sha256(data.encode()).hexdigest()
    print(data, "→", h[:10])
    prev = h
```
